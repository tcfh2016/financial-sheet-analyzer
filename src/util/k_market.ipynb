## 北上资金

```python
import akshare as ak

stock_hsgt_north_net_flow_in_em_df = ak.stock_hsgt_north_net_flow_in_em(symbol="北上")
stock_hsgt_north_net_flow_in_em_df.set_index(['date'], inplace=True)
stock_hsgt_north_net_flow_in_em_df = stock_hsgt_north_net_flow_in_em_df[:]
stock_hsgt_north_net_flow_in_em_df.plot(figsize=(10, 5))
print(stock_hsgt_north_net_flow_in_em_df)

stock_hsgt_north_acc_flow_in_em_df = ak.stock_hsgt_north_acc_flow_in_em(symbol="北上")
stock_hsgt_north_acc_flow_in_em_df.set_index(['date'], inplace=True)
stock_hsgt_north_acc_flow_in_em_df = stock_hsgt_north_acc_flow_in_em_df[:]
stock_hsgt_north_acc_flow_in_em_df.plot(figsize=(10, 5))
print(stock_hsgt_north_acc_flow_in_em_df)
```
